```python
group = df.groupby('INVOICE_NO').agg({'INVOICE_NO': lambda x: x.nunique(),
                                            'PRODUCT':lambda x: x,
                                            'QUANTITY':lambda x: x.sum(),
                                            'RATE':lambda x:x.sum(),
                                            'AMOUNT_AFTER_TAX':lambda x:x,
                                            })
group.columns = ['INVOICE_NO','PRODUCT','QUANTITY','RATE','AMOUNT_AFTER_TAX']
```

```python
group
```

```python
# ITEM QUANTITY TOP 10

top_10 = df.groupby('PRODUCT')['QUANTITY'].sum().nlargest(10)

sns.barplot(x=top_10.values,y=top_10.index)
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top_10')
plt.xticks(rotation=50)
plt.show()
```

```python
top_10_amt = df.groupby('PRODUCT')['AMOUNT_AFTER_TAX'].sum().nlargest(10)
sns.barplot(x=top_10_amt.values,y=top_10_amt.index)
plt.xlabel('Count')
plt.ylabel('Product')
plt.title('Top_10_amt')
plt.xticks(rotation=50)
plt.show()
```
